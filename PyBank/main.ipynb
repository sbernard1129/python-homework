```python
import numpy as np
import math
import csv
from pathlib import Path

# csvpath = Path("../Resources/test.csv")
csvpath = Path("../Resources/budget_data.csv")

# intialize variables
date_count = 0
total_pnl = 0
max_increase = 0
max_decrease = 0
sum_of_changes = 0
max_increase_date = ""
max_decrease_date = ""

pnl_list = [0]
change_list = []

# Open the csv file as an object
with open(csvpath, 'r') as csvfile:
    
    # Create reader object with comma as delimiter
    csvreader = csv.reader(csvfile, delimiter=",")
    
    # Store csv header
    csv_header = next(csvreader)

    for row in csvreader:
        
        # create variables date and pnl. Clean up the string for date and cast pnl as float
        date = str(row[0].strip())
        pnl = float(row[1]) 
        
        # if date is a non-empty string, increment date count. Otherwise, we have a missing date and ignore the row
        if(bool(date)):
            
            # append pnl to pnl_list
            pnl_list.append(pnl)
            
            # The change is the pnl on this loop iteration minus the last iteration. On first loop, we store the pnl instead to be popped off later
            change_list.append(pnl - pnl_list[-2])
            
            # changes are summed up to later calculate mean
            sum_of_changes += change_list[-1]
            
            # total and date count are increased
            date_count += 1            
            total_pnl += pnl
            
            if(change_list[-1] > max_increase and date_count > 1):
                max_increase = change_list[-1]
                max_increase_date = date
            elif(change_list[-1] < max_decrease and date_count > 1):
                max_decrease = change_list[-1]
                max_decrease_date = date

# if we at least one record
if(date_count > 0):
    # Remove initial value from sum, which is not a change but the first record of pnl 
    sum_of_changes -= change_list[0]
    
    # Remove leading zero from PnL list
    pnl_list.pop()
    
    # Remove filler starter value from changes list
    change_list.pop()
    
    
    if(date_count == 1):
        average_change = 0
        increase = "None"
        decrease = "None"
    
    elif(date_count != 0):
        average_change = sum_of_changes / (date_count - 1)
        increase = f"{max_increase_date} (${max_increase:.2f})"
        decrease = f"{max_decrease_date} (${max_decrease:.2f})"
        
    output = f"Financial Analysis\n----------------------------\nTotal Months: {date_count}\nTotal: ${total_pnl:.2f}\nAverage Change: ${average_change:.2f}\n"
    output += f"Greatest Increase in Profits: {increase}\nGreatest Decrease in Profits: {decrease}"

    print(f"{output}")
    
    print(f"\nWriting to output.txt ...")
    
    output_path = 'output.txt'

    # Open the output path as a file object
    with open(output_path, 'w') as file:
        # Write daily_average to the output file, convert to string
        file.write(f"{output}")

# else no records
else:
    print(f"ERROR: no records found")
```
